```python
import matplotlib.pyplot as plt
```

# Load data

```python
df = pd.read_csv("wines.csv")
```

```python
df
```

# How many wines have been given a rating of 100 points ?

```python
len(df.loc[df["points"] == 100])
```

# What is the name of the most expansive wines ?

```python
df.loc[df["price"] == df["price"].max()]["name"].squeeze()
```

# Calculate a new column where you show the rating in a scale of 0 to 5 Floats are allowed

```python
df["new"] = df["points"] / 20
df["new"]
```

# Create a price histogram from wines that cost less than 100

```python
df.loc[df["price"] < 100["price"].hist()
```

# Plot price horizontally against points vertically. Do you think ratings increase by price ? Yes

```python
df.plot(x="price", y="points", figsize=(22, 3), kind="scatter")
```
